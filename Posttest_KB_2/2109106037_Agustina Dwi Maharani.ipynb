##### dictionary tipe data

```python
dicts = {"object":0, "int64":0, "float64":0}

for i in df.columns:
    if df[i].dtypes == "object":
        dicts["object"] += 1
    if df[i].dtypes == "int64":
        dicts['int64'] += 1
    if df[i].dtypes == "float64":
        dicts['float64'] += 1
        
print(dicts)
```
