```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

### Grafik Pie

```python
df =pd.read_csv("healthcare_dataset_stroke_data.csv")
df.head()
```

```python
data_pie = df["work_type"].value_counts()
data_pie.head()
```

```python
data_pie = df["work_type"].value_counts().rename_axis("work_type")
data_pie.head()
```

```python
data_pie = df["work_type"].value_counts().rename_axis("work_type").reset_index(name="Jumlah")
data_pie.head()
```

### Menampilkan Presentase Kategori Pekerjaan berdasarkan populasinya dengan grafik pie

```python
plt.pie(data_pie["Jumlah"],
        labels=data_pie["work_type"],
        explode=(0.03,0.03,0.03,0.03,0.03), 
        startangle=90, 
        autopct='%.2f%%')

plt.show()
```

## Scatter Plot

### Menampilkan bmi berdasarkan id dan residence_type dengan scatter plot

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10,5))
sns.scatterplot(data=df,x="bmi",y="id",hue="Residence_type")

plt.xlabel('BMI')
plt.ylabel('ID')

plt.show()
```
